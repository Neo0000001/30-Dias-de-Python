2. Use for loop para iterar de 0 a 100 e imprimir la suma de todos los pares y la suma de todos los impares.

    The sum of all evens is 2550. And the sum of all odds is 2500.

````python
suma_pares = 0
suma_impares = 0

for i in range(101):
    if i % 2 == 0:
        suma_pares += i
    else:
        suma_impares += i

print('La suma de los numeros pare es:', suma_pares)
print('La suma de los numeros impares es:', suma_impares)
````

2. Esta es una lista de frutas, ['plátano', 'naranja', 'mango', 'limón'] invierta el orden usando bucle.

````python
frutas = ['plátano', 'naranja', 'mango', 'limón']

for i in range(int(len(frutas)/2)):
    frutas[i], frutas[len(frutas)-1-i] = frutas[len(frutas)-1-i], frutas[i]

print(frutas)
````

Otra forma de hacerlo es usando el método reverse()

````python
frutas = ['plátano', 'naranja', 'mango', 'limón']
frutas.reverse()
print(frutas)
````
